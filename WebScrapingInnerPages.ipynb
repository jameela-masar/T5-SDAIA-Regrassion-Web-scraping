# import libraries

```python
from bs4 import BeautifulSoup as bs
import requests
import numpy as np
import pandas as pd
import csv
from itertools import zip_longest
import matplotlib.pyplot as plt
%matplotlib inline
```

# Read the data

```python
main_df = pd.read_csv("part5.csv")
main_df.head()
```

# Try , Except

```python
def get_data(team, i):
    try:
        return team.find_all("td")[i].text
    except:
        return np.nan
```

```python
import time
time.sleep(4)
```

# Read data from links

```python
pl_teams=[]
a=[]

page_dict = {
    "Area":[],
    "Front":[],
    "Bedrooms":[],
    "Salon":[],
    "Bathrooms":[],
    "Street_width":[],
    "Age":[]
}
for l in main_df.Link:
    result=requests.get(l)
    soup = bs(result.text, "lxml")
    table=soup.find('table',class_='dc-detailsTable')
    for team in table.find_all('tbody'):
        page_dict["Area"].append(get_data(team, 0))
        page_dict["Front"].append(get_data(team, 2))
        page_dict["Bedrooms"].append(get_data(team, 4))
        page_dict["Salon"].append(get_data(team, 6))
        page_dict["Bathrooms"].append(get_data(team, 8))
        page_dict["Street_width"].append(get_data(team, 10))
        page_dict["Age"].append(get_data(team, 12))
        
    time.sleep(4)
    
        
         
sub_df = pd.DataFrame(page_dict)
```

```python
sub_df.head()
```

# Merge DataFrame

```python
House1=main_df.merge(sub_df, left_index= True, right_index=True)

House1
```
